```python
def preprocess_text(filepath, ext):
    with open(f'./{filepath}.{ext}', 'r') as f:
        text = f.read()

        text = text.split('\r\n')[0].split('\n')

        text = ' '.join(text)
    
    return text
```

```python
lyrics = preprocess_text('yah_mean', 'txt')
print(lyrics)
```

```python
def make_markov_chain(lyrics, n):
    n_gram_hash = {}
    for i in range(len(lyrics) - (n+1)):
        curr_gram = lyrics[i: i+n]
        
        if curr_gram not in n_gram_hash.keys():
            n_gram_hash.update({curr_gram: []})
        
        arr = n_gram_hash[curr_gram]
        arr.append(lyrics[i+n])
        n_gram_hash[curr_gram] = arr
    
    return n_gram_hash
```

```python
import random

def generate_lyrics(ngram_hash, starting, order):
    curr_gram = starting
    generated_lyrics = curr_gram
    
    for i in range(100):
        next_chars = ngram_hash[curr_gram]
        rand_next = next_chars[random.randint(0, len(next_chars)-1)]
        
        generated_lyrics += rand_next
        
        curr_gram = generated_lyrics[len(generated_lyrics)-order: len(generated_lyrics)]
    
    return generated_lyrics
```

```python
order = 4
ngram_hash = make_markov_chain(lyrics, order)
print(ngram_hash)
```

```python
for i in range(10):
    lyrics = generate_lyrics(ngram_hash, 'All ', order)
    print(lyrics)
```
